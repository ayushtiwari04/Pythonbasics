 Q:2- Explain the difference between mutable and immutable data types with examples?

 -: MUTABLE DATATYPE:

 Objects whose values can be changed/modified after creation without changing their identity (memory location).Examples{List,Dictionary,Set}.

 IMMUTABLE DATATYPES:

Q:5- Why might you use a set instead of a list in Python?

ANS:use a set instead of a list when you want unique elements and fast membership tests, especially if you plan to do math-like operations on collections.

Reasons to use a set instead of a list:

Automatic Duplicate Removal
Sets cannot contain duplicates.
Example:
my_list = [1, 2, 2, 3]
my_set = set(my_list)
print(my_set)   # {1, 2, 3}
Set Operations (Math-like)
Sets support union, intersection, difference, and other operations that lists dont natively support.
Example:
a = {1, 2, 3}
b = {3, 4, 5}
print(a & b)  # Intersection → {3}
print(a | b)  # Union → {1, 2, 3, 4, 5}

Q:14- How does the “in” keyword work differently for lists and dictionaries?

-: For lists, "in" checks if a value exists among the elements.

For dictionaries, "in" checks if a key exists (not values).
d = {"a": 1, "b": 2}
"a" in d   # True
1 in d     # False

Q:19- Why are dictionaries considered unordered, and how does that affect data retrieval?

-: A dictionary is often referred to as an unordered collection of objects because the elements within a dictionary are not arranged in a specific order.
How this affects data retrieval:-

1)Direct access is order-independent.You retrieve values by their key, not by position.Example:

my_dict = {"apple": 1, "banana": 2, "cherry": 3}
print(my_dict["banana"])  

2)When you loop through a dictionary, you now get items in insertion order (Python 3.7+). But logically, you shouldn't rely on the order unless you explicitly need it.
Example:

for key in my_dict:
    print(key)

Q:20- Explain the difference between a list and a dictionary in terms of data retrieval?

-: LIST:-Use a list when order matters and you want to retrieve elements by their position.

Retrieval method: By index (position).

Example:

my_list = ["apple", "banana", "cherry"]
print(my_list[1])   # "banana"

DICTIONARY:-Use a dictionary when you want to retrieve elements by a meaningful key quickly, regardless of order.
Retrieval method: By key (mapping).

Example:

my_dict = {"apple": 1, "banana": 2, "cherry": 3}
print(my_dict["banana"])   # 2
